## 3. Models

### DNS model and Eurodollar Futures Rates
First, we regard ED Futures Rates as forward rates.
Then:
$$ 1+\frac{ED \space rates}{4} = e^{y(t,T_2)\times(T_2-t)-y(t,T_1)\times(T_1-t)}$$
where $y(t,T_2), y(t,T_1)$ is the zero coupon bond yield to maturity $T_1,T_2$ at time t respectively;
the $T_1$ is the IMM date for this futures contract, while $T_2 = T_1 + 3 \space month$   
The DNS model shows that:
$$ y(t,T) = \beta_{0t} + \beta_{1t} \times \frac{1-e^{-\lambda(T-t)}}{\lambda(T-t)} +\beta_{2t} \times (\frac{1-e^{-\lambda(T-t)}}{\lambda(T-t)} - e^{-\lambda(T-t)})$$

### Optimization Problem
From above, we can derive that:
$$ln(1+\frac{ED\space rates}{4}) = \beta_{0t}\times (T_2 - T_1) + \beta_{1t} \times \frac{e^{-\lambda(T_1-t)}-e^{-\lambda(T_2-t)}}{\lambda} +\beta_{2t} \times (\frac{e^{-\lambda(T_1-t)}-e^{-\lambda(T_2-t)}}{\lambda}+(T_1-t)\times e^{-\lambda \times (T_2-t)}-(T_2-t)\times e^{-\lambda(T_2-t)})$$
Suppose:
$$ED^{mkt} = ln(1+\frac{ED\space rates}{4}) $$
$$ED^{model} = \beta_{0t}\times (T_2 - T_1) + \beta_{1t} \times \frac{e^{-\lambda(T_1-t)}-e^{-\lambda(T_2-t)}}{\lambda} +\beta_{2t} \times (\frac{e^{-\lambda(T_1-t)}-e^{-\lambda(T_2-t)}}{\lambda}+(T_1-t)\times e^{-\lambda \times (T_1-t)}-(T_2-t)\times e^{-\lambda(T_2-t)})$$
Then the error will be:
$$e = ED^{mkt} - ED^{model}$$

```python
mkt = np.log(1+ED_rates/4.)
```

```python
mkt.head()
```

```python
lamb0 = 0.0400 # The fixed lambda for Model B
```

## 4. Model A Estimation

SSR Function:
* exog: $X$ matrix in linear regression
* b: fitted betas, the formula is $\beta = (X^{T}X)^{-1}X^{T}y$
* e: errors, the formula is $error = y-\tilde{y} = y - X\beta$

```python
def SSR(lam):
    ssr = 0
    cols = mkt.columns
    for j in mkt.index:
        m = [TTMs_d.loc[j,i] for i in cols]
        exog = np.array([[3,(math.exp(-lam*i)-math.exp(-lam*(i+3)))/lam,((math.exp(-lam*i)-math.exp(-lam*(i+3)))/lam+i*math.exp(-lam*i)-(i+3)*math.exp(-lam*(i+3)))] for i in m], dtype='float')
        b = np.dot(np.linalg.inv(exog.T.dot(exog)), exog.T.dot(mkt.loc[j,:]))
        e = mkt.loc[j,:]-exog.dot(b)
        ssr = ssr + e.T.dot(e)
    return ssr
```

```python
import scipy.optimize as sp_optimize
```

```python
optimize = sp_optimize.minimize(SSR,lamb0)
```

```python
optimize
```

* So the optimal $\lambda$ for Model A is **0.06008374**

```python
ModelA_factors = pd.DataFrame(columns = ['b1','b2','b3'])
lamb = optimize['x'][0]
cols = mkt.columns
for j in mkt.index:
    m = [TTMs_d.loc[j,i] for i in cols]
    exog = np.array([[3,(math.exp(-lamb*i)-math.exp(-lamb*(i+3)))/lamb,((math.exp(-lamb*i)-math.exp(-lamb*(i+3)))/lamb+i*math.exp(-lamb*i)-(i+3)*math.exp(-lamb*(i+3)))] for i in m], dtype='float')
    ModelA_factors.loc[j,:] = np.dot(np.linalg.inv(exog.T.dot(exog)), exog.T.dot(mkt.loc[j,:]))
```

```python
ModelA_factors.head()
```

## 5. Check Collinearity for Model A
* In this section, we will check the condition numbers for everyday. The large condition number means collinearity, then we will apply PCA for remedy.
* Condition number: 
For a linear regression, if $X$ is the design matrix, then the condition number will be:
$$ \kappa(X^T X) = \frac{\lambda_{max}(X^TX)}{\lambda_{min}(X^TX)}$$

```python
for j in ModelA_factors.index:
    m = [TTMs_d.loc[j,i] for i in cols]
    exog = np.array([[3,(math.exp(-lamb*i)-math.exp(-lamb*(i+3)))/lamb,((math.exp(-lamb*i)-math.exp(-lamb*(i+3)))/lamb+i*math.exp(-lamb*i)-(i+3)*math.exp(-lamb*(i+3)))] for i in m], dtype='float')
    c = exog.T.dot(exog)
    val, vec = np.linalg.eig(c)
    ModelA_factors.loc[j,'condition number'] = val.max()/val.min()
```

```python
ModelA_factors.head()
```

* From above, the condition numbers are always more than 100, so it might be a little collinear.

## 5. Continue: Collinearity Remedy for Model A
* In this section, we use ridge regression to eliminate the collinearity.

```python
ModelA_factors2 = pd.DataFrame(columns = ['b1','b2','b3'])
k = 1
for j in mkt.index:
    m = [TTMs_d.loc[j,i] for i in cols]
    exog = np.array([[3,(math.exp(-lamb*i)-math.exp(-lamb*(i+3)))/lamb,((math.exp(-lamb*i)-math.exp(-lamb*(i+3)))/lamb+i*math.exp(-lamb*i)-(i+3)*math.exp(-lamb*(i+3)))] for i in m], dtype='float')
    ModelA_factors2.loc[j,['b1','b2','b3']] = np.dot(np.linalg.inv(exog.T.dot(exog)+k*np.diag(np.ones(3))), exog.T.dot(mkt.loc[j,:]))
    c = exog.T.dot(exog)+k*np.diag(np.ones(3))
    val, vec = np.linalg.eig(c)
    ModelA_factors2.loc[j,'condition number'] = val.max()/val.min()
```

```python
ModelA_factors2.head()
```

* From above, we set the k value as 1, so that the new condition number is less than 100.

## 6. Model B Estimation

```python
ModelB_factors = pd.DataFrame(columns = ['b1','b2','b3'])
for j in mkt.index:
    m = [TTMs_d.loc[j,i] for i in cols]
    exog = np.array([[3,(math.exp(-lamb0*i)-math.exp(-lamb0*(i+3)))/lamb0,((math.exp(-lamb0*i)-math.exp(-lamb0*(i+3)))/lamb0+i*math.exp(-lamb0*i)-(i+3)*math.exp(-lamb0*(i+3)))] for i in m], dtype='float')
    ModelB_factors.loc[j,:] = np.dot(np.linalg.inv(exog.T.dot(exog)), exog.T.dot(mkt.loc[j,:]))
```

```python
for j in ModelB_factors.index:
    m = [TTMs_d.loc[j,i] for i in cols]
    exog = np.array([[3,(math.exp(-lamb0*i)-math.exp(-lamb0*(i+3)))/lamb0,((math.exp(-lamb0*i)-math.exp(-lamb0*(i+3)))/lamb0+i*math.exp(-lamb0*i)-(i+3)*math.exp(-lamb0*(i+3)))] for i in m], dtype='float')
    c = exog.T.dot(exog)
    val, vec = np.linalg.eig(c)
    ModelB_factors.loc[j,'condition number'] = val.max()/val.min()
```

```python
ModelB_factors2 = pd.DataFrame(columns = ['b1','b2','b3'])
k = 1.6
for j in mkt.index:
    m = [TTMs_d.loc[j,i] for i in cols]
    exog = np.array([[3,(math.exp(-lamb0*i)-math.exp(-lamb0*(i+3)))/lamb0,((math.exp(-lamb0*i)-math.exp(-lamb0*(i+3)))/lamb0+i*math.exp(-lamb0*i)-(i+3)*math.exp(-lamb0*(i+3)))] for i in m], dtype='float')
    ModelB_factors2.loc[j,['b1','b2','b3']] = np.dot(np.linalg.inv(exog.T.dot(exog)+k*np.diag(np.ones(3))), exog.T.dot(mkt.loc[j,:]))
    c = exog.T.dot(exog)+k*np.diag(np.ones(3))
    val, vec = np.linalg.eig(c)
    ModelB_factors2.loc[j,'condition number'] = val.max()/val.min()
```
